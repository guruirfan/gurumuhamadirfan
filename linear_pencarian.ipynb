```python
# Fungsi linear_search untuk mencari data tertentu dalam sebuah list
def linear_search(data, cari):
    indek = len(data)      # panjang list data
    value = False          # variabel penanda apakah data ditemukan atau tidak
    data1 = []             # list untuk menyimpan indeks data yang ditemukan

    # loop dari 0 sampai panjang data
    for i in range(0, indek):
        if cari == data[i]:   # cek apakah elemen data sama dengan nilai yang dicari
            value = True      # jika ketemu, ubah penanda jadi True
            data1.append(i)   # simpan indeks elemen yang ditemukan

    # setelah looping selesai, cek apakah data ditemukan
    if value == True:
        print("data ketemu :")
        for i in data1:
            print("di indek : ", i)   # cetak indeks tempat data ditemukan
    else:
        print("data tidak ada")       # jika tidak ditemukan

# data list yang akan dicari
data = [4, 3, 5]   # kamu bisa isi list ini dengan angka lain
print("data : ", data)

# input dari pengguna untuk nilai yang ingin dicari
cari = int(input("cari data apa : "))
linear_search(data, cari)   # panggil fungsi untuk mencari data
```
